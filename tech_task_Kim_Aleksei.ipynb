```python
import pandas as pd 
import plotly.graph_objects as go
import scipy.stats as stats
from scipy.stats import ttest_ind

source = pd.read_csv(filepath_or_buffer = 'data/2nd_task.csv')

source.head()
```

```python
source.info()
```

Для начала отфильтруем исходный dataframe, оставив только те строки, которые соответствуют категории товара (Category) = Пицца, так как данный тест был связан только со страницей каталога пиццы

Также отфильтруем исходный dataframe, оставив только строки, которые по источнику товара относятся только к Web-сайту, опять же потому что данный тест проводился только для Web-сайта

Стоит также ограничить датасэт по датам и оставить данные только за те даты, которые попали во временной промежуток проведения теста


```python
source = source[source['Category'] == 'пицца']

source = source[source['OrderSource'] == 'web']

source['Date'] = pd.to_datetime(source['Date'], format='%Y-%m-%d')

source = source.sort_values(by='Date')

#обозначаем период теста
start_date = '2022-04-07'
end_date = '2022-05-19'

# Filter the DataFrame by the date range
source = source[(source['Date'] >= start_date) & (source['Date'] <= end_date)]

uniq_dates = source['Date'].unique()

print(source, uniq_dates)
```

Также исходный dataframe будет разделен на 2 dataframe-а:

Первый df будет содержать данные по городам, в которых проводился тест (test_df)

Второй df будет содержать данные по остальным городам (control_df)

```python
#Выведем все уникальные названия городов, которые присутствуют в dataframe
uniq_cities = source['CityName'].unique()

print(uniq_cities)
```

```python
#формируем лист из городов, в которых проводился тест
cities_for_test = ['Пенза','Уфа','Курск','Нижний Тагил','Новокуйбышевск','Орск']

#отфильтровываем строки dataframe по городам для теста
test_df = source[source['CityName'].isin(cities_for_test)]

test_df.head()
```

```python
#формируем лист из городов, которые не участвовали в тесте
cities_for_control = [city for city in uniq_cities if city not in cities_for_test]

#формируем лист из городов, в которых НЕ проводился тест (контрольная группа)
control_df = source[source['CityName'].isin(cities_for_control)]

control_df.head()
```

```python
#меняем тип данных в колонках дат с object на datetime и меняем формат даты в обоих df
test_df['Date'] = pd.to_datetime(test_df['Date'])


control_df['Date'] = pd.to_datetime(control_df['Date'])


#выведем колонку выручки, которая будет посчитана как количество заказов * средняя цена

test_df['rev'] = test_df['tickets'] * test_df['avg_price']

control_df['rev'] = control_df['tickets'] * control_df['avg_price']

test_df.head()
```

```python
#формируются DF для визуализации временных рядов
revenue_by_dt_tst =  test_df.groupby('Date', as_index=False).rev.sum()

revenue_by_dt_tst['Date'] = pd.to_datetime(revenue_by_dt_tst['Date'])

revenue_by_dt_tst = revenue_by_dt_tst.sort_values(by='Date')


revenue_by_dt_ctrl =  control_df.groupby('Date', as_index=False).rev.sum()

revenue_by_dt_ctrl['Date'] = pd.to_datetime(revenue_by_dt_tst['Date'])

revenue_by_dt_ctrl = revenue_by_dt_ctrl.sort_values(by='Date')

uniq_dates_ctrl = revenue_by_dt_ctrl['Date'].unique()
```

Анализ динамики показателей, в контрольной и тестовой dataframe-ах, используя plotly 

```python
#создается словарь со стилизацией графиков для plotly
axis_props = dict(showgrid = True,
                gridcolor = '#696969',
                gridwidth = 0.5
                )

#создается экземпляр класса Figure из plotly.go для выручки из тестовой группы
dynamic_rev_tst = go.Figure()

dynamic_rev_tst.add_trace(go.Scatter(x=revenue_by_dt_tst['Date'],
                             y=revenue_by_dt_tst['rev'],
                             mode='lines',
                             name='Line 1'))


dynamic_rev_tst.update_layout(title = 'Динамика выручки в тестовой группе',
                            plot_bgcolor = '#1e1e1e',
                            paper_bgcolor = '#1e1e1e',
                            title_font_color="#fff",
                            font_color = '#fff',
                            title_font_size = 20,
                            xaxis = axis_props,
                            yaxis = axis_props,
                            width = 1550,
                            height = 600)


#создается экземпляр класса Figure из plotly.go для выпручки из контрольной группы
dynamic_rev_ctrl = go.Figure()

dynamic_rev_ctrl.add_trace(go.Scatter(x=revenue_by_dt_ctrl['Date'],
                             y=revenue_by_dt_ctrl['rev'],
                             mode='lines',
                             name='Line 1'))


dynamic_rev_ctrl.update_layout(title = 'Динамика выручки в контрольной группе',
                            plot_bgcolor = '#1e1e1e',
                            paper_bgcolor = '#1e1e1e',
                            title_font_color="#fff",
                            font_color = '#fff',
                            title_font_size = 20,
                            xaxis = axis_props,
                            yaxis = axis_props,
                            width = 1550,
                            height = 600)

#выводятся два графика
dynamic_rev_tst.show()

dynamic_rev_ctrl.show()
```

Отсюда выдвигается первая гипотеза: средняя выручка в обоих группах не отличается значительно

Либо клиенты начали изучать новые виды пицц, исходя из их рейтинга, что привело к увеличению времени одной сессии при заказе

Статистическая проверка гипотез

H0: Средняя цена заказа в контрольной группе больше или равна средней цене в тестовой группе

H1: Средняя цена заказа в контрольной группе меньше средней цены в тестовой группе

Примем уровень значимости, равный 0,05

```python
t_stat, p_value = ttest_ind(dist_tickets_and_price_ctrl['avg_price'],
                            dist_tickets_and_price_tst['avg_price'],
                            alternative='less')

print(t_stat, p_value)
```

P value больше принятого уровня значимости, что означает, что нулевая гипотеза отвергается

То есть, Средняя цена заказа в контрольной группе меньше средней цены в тестовой группе 

H0: Среднее время сессии в контрольной группе больше или равна среднему времени сессии в тестовой группе

H1: Среднее время сессии в контрольной группе меньше среднего времени сессии в тестовой группе

Примем уровень значимости, равный 0,05

```python
t_stat, p_value = ttest_ind(dist_tickets_and_session_ctrl['avg_time'],
                            dist_tickets_and_session_tst['avg_time'],
                            alternative='less')

print(t_stat, p_value)
```

P value больше принятого уровня значимости, что означает, что нулевая гипотеза отвергается

То есть, cреднее время сессии в контрольной группе меньше среднего времени сессии в тестовой группе
